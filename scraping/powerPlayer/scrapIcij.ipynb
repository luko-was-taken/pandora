Recupére des infos (nom, fonction et lien de la story) des power players (politique)

```python
from bs4 import BeautifulSoup as bs
import requests

url="https://offshoreleaks.icij.org/stories"

page = requests.get(url)
print(page.text)

soup = bs(page.content, "html.parser")
soup
```

```python
listSearchPowerPlayers = soup.findAll("div", class_="pure-u-1 pure-u-lg-1-3 pure-u-sm-1-2")
print(listSearchPowerPlayers)
```

```python
tabPowerPlayers = []
tabDoc=[]
tabArt=[]
tabFunctionAndCountryPowerPlayer = []
i = 1
for element in listSearchPowerPlayers:

    #-----recupére le nom du power player------  
    namepowerPlayer = element.find("div", class_="stories_results_cont_list_item_title").get_text().strip()
    print(namepowerPlayer)

    #-----recupére sa fonction et son pays------
    functionAndCountryPowerPlayer = element.find("h3", class_="stories_results_cont_list_item_name").get_text().strip()
    
    tabFunctionAndCountryPowerPlayer = functionAndCountryPowerPlayer.split(",")
    functionPowerPlayer = tabFunctionAndCountryPowerPlayer[0]
    print(functionPowerPlayer)  
    if(len(tabFunctionAndCountryPowerPlayer) == 2):
        countryPowerPlayer = tabFunctionAndCountryPowerPlayer[1].lstrip()      
        print(countryPowerPlayer)    
      
    #-----recupére les documents------
    linkStoryPowerPlayer = element.find("a", class_="stories_results_cont_list_item", href=True)
    print(linkStoryPowerPlayer["href"])
    storyPageUrl = "https://offshoreleaks.icij.org"+ linkStoryPowerPlayer["href"]

    pageStory = requests.get(storyPageUrl)
  
    soupStory = bs(pageStory.content, "html.parser")  

    heading = soupStory.findAll("h3")
   
    n=len(heading)
    for x in range(n):
        if str.strip(heading[x].text) == "Related documents":
            targetLien = heading[x].parent       
            linkDoc = targetLien.findAll("a", href=True)
            for idoc in linkDoc:  
                tabDoc.append(idoc["href"])  
                print(idoc["href"] + "\n")
                

        #-----recupére les articles------
        if str.strip(heading[x].text) == "Related articles":
            targetLienArt = heading[x].parent       
            linkArt = targetLienArt.findAll("a", href=True)
            for iArt in linkArt:
                print(iArt["href"] + "\n")
                tabArt.append(iArt["href"])
    #-----insérer les données dans un tableau------         
    tabPowerPlayers.append({"id": i, "nom":namepowerPlayer,"fonction": functionPowerPlayer,"pays": countryPowerPlayer ,"document":[idD for idD in tabDoc],"article": [idA for idA in tabArt]})    
    i+=1
    
```

```python
#j'importe mon tableau dans un fichier jSON
import json

jsonString = json.dumps(tabPowerPlayers)
jsonFile = open("dataPowerPlayer.json", "w")
jsonFile.write(jsonString)
jsonFile.close()
```
